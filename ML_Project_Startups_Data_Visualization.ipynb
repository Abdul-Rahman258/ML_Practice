## Data cleaning

### Subtask:
Clean the data by handling potential outliers in the numerical features.


**Reasoning**:
I will identify outliers in the numerical features using the IQR method, define upper and lower bounds, replace outliers with the bounds, and then re-examine the distributions using box plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate IQR for numerical features
numerical_features = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']
for feature in numerical_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    # Replace outliers with bounds
    df[feature] = df[feature].clip(lower=lower_bound, upper=upper_bound)

# Re-examine distributions using box plots
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Distribution of {col} after outlier handling')
plt.tight_layout()
plt.show()
```

## Model optimization

### Subtask:
Optimize the hyperparameters of the three trained regression models (Linear Regression, Random Forest, and Gradient Boosting) using GridSearchCV. Evaluate their performance on the test set using R-squared, Mean Squared Error, and Mean Absolute Error.


**Reasoning**:
Import necessary libraries and perform hyperparameter tuning using GridSearchCV for the three regression models.



```python
# Define parameter grids for each model
param_grid_linear = {'fit_intercept': [True, False], 'positive': [True, False]} # Corrected parameter grid for linear model
param_grid_rf = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10]}
param_grid_gb = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1], 'max_depth': [3, 5, 7]}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, r2_score, mean_squared_error, mean_absolute_error
import numpy as np

# Define parameter grids for each model
param_grid_linear = {'fit_intercept': [True, False], 'positive': [True, False]} #Corrected parameter grid for linear model
param_grid_rf = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10]}
param_grid_gb = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1], 'max_depth': [3, 5, 7]}

# Create scoring metrics
scoring = {'r2': make_scorer(r2_score),
           'mse': make_scorer(mean_squared_error, greater_is_better=False),
           'mae': make_scorer(mean_absolute_error)}

# Create GridSearchCV objects
grid_linear = GridSearchCV(linear_model, param_grid_linear, scoring=scoring, refit='r2', cv=5)
grid_rf = GridSearchCV(rf_model, param_grid_rf, scoring=scoring, refit='r2', cv=5)
grid_gb = GridSearchCV(gb_model, param_grid_gb, scoring=scoring, refit='r2', cv=5)

# Fit GridSearchCV objects to the training data
grid_linear.fit(X_train_numeric, y_train)
grid_rf.fit(X_train_numeric, y_train)
grid_gb.fit(X_train_numeric, y_train)

# Evaluate best models on the test data
y_pred_linear = grid_linear.best_estimator_.predict(X_test_numeric)
y_pred_rf = grid_rf.best_estimator_.predict(X_test_numeric)
y_pred_gb = grid_gb.best_estimator_.predict(X_test_numeric)

# Calculate evaluation metrics
r2_linear = r2_score(y_test, y_pred_linear)
mse_linear = mean_squared_error(y_test, y_pred_linear)
mae_linear = mean_absolute_error(y_test, y_pred_linear)

r2_rf = r2_score(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)

r2_gb = r2_score(y_test, y_pred_gb)
mse_gb = mean_squared_error(y_test, y_pred_gb)
mae_gb = mean_absolute_error(y_test, y_pred_gb)

# Print results
print("Linear Regression:")
print("Best Hyperparameters:", grid_linear.best_params_)
print(f"R-squared: {r2_linear:.4f}, MSE: {mse_linear:.4f}, MAE: {mae_linear:.4f}")

print("\nRandom Forest:")
print("Best Hyperparameters:", grid_rf.best_params_)
print(f"R-squared: {r2_rf:.4f}, MSE: {mse_rf:.4f}, MAE: {mae_rf:.4f}")

print("\nGradient Boosting:")
print("Best Hyperparameters:", grid_gb.best_params_)
print(f"R-squared: {r2_gb:.4f}, MSE: {mse_gb:.4f}, MAE: {mae_gb:.4f}")
```

```python
# Evaluate the models based on R-squared, MSE, and MAE
models = {
    'Linear Regression': (r2_linear, mse_linear, mae_linear),
    'Random Forest': (r2_rf, mse_rf, mae_rf),
    'Gradient Boosting': (r2_gb, mse_gb, mae_gb),
}

best_model = None
best_r2 = -1  # Initialize with a value less than the minimum possible R-squared
for model_name, (r2, mse, mae) in models.items():
    print(f'{model_name}: R-squared = {r2:.4f}, MSE = {mse:.4f}, MAE = {mae:.4f}')
    if r2 > best_r2:
        best_r2 = r2
        best_model = model_name
    elif r2 == best_r2:
      # if R2 is same then select the model based on MSE or MAE
      if model_name == 'Linear Regression':
        best_model = 'Linear Regression'
      elif mse < mse_linear:
        best_model = model_name
      elif mse == mse_linear:
        if mae < mae_linear:
          best_model = model_name

if best_model:
    print(f"\nBest Model: {best_model}")
    print(f"Reasoning: {best_model} has the highest R-squared value of {best_r2:.4f}")
else:
    print("\nNo suitable model found based on the evaluation metrics.")
```

## Data preparation

### Subtask:
Prepare the best performing model (Linear Regression) for user interaction.


**Reasoning**:
Save the best performing model and create a function to interact with the model.



```python
import pickle

# Save the best model
filename = 'best_model.pkl'
pickle.dump(grid_linear.best_estimator_, open(filename, 'wb'))

def predict_profit(rd_spend=None, administration=None, marketing_spend=None, year=None, state=None, total_spend=None, total_spend_year=None, company_name=None, input_feature=None):
    """
    Predicts profit based on the provided input features.

    Args:
        rd_spend (float): R&D spend.
        administration (float): Administration cost.
        marketing_spend (float): Marketing spend.
        year (int): Year.
        state (str): State ('California', 'Florida', or 'New York').
        total_spend (float): Total spend.
        total_spend_year (float): Total spend * Year.
        company_name (str): Company Name.
        input_feature (str): The input feature to predict others.


    Returns:
        dict or None: A dictionary containing predicted profit and other monetary values, or None if input is invalid.
    """

    try:
        # Load the saved model
        loaded_model = pickle.load(open(filename, 'rb'))

        # Create input array
        input_data = pd.DataFrame(columns=X_train_numeric.columns)

        # Handle state input
        if state == 'California':
            input_data['State_California'] = [1]
            input_data['State_Florida'] = [0]
            input_data['State_New York'] = [0]
        elif state == 'Florida':
            input_data['State_California'] = [0]
            input_data['State_Florida'] = [1]
            input_data['State_New York'] = [0]
        elif state == 'New York':
            input_data['State_California'] = [0]
            input_data['State_Florida'] = [0]
            input_data['State_New York'] = [1]
        else:
            return None

        if input_feature == "R&D Spend":
            input_data['R&D Spend'] = [rd_spend]
            input_data['Administration'] = [administration]
            input_data['Marketing Spend'] = [marketing_spend]
        elif input_feature == "Administration":
            input_data['R&D Spend'] = [rd_spend]
            input_data['Administration'] = [administration]
            input_data['Marketing Spend'] = [marketing_spend]
        elif input_feature == "Marketing Spend":
            input_data['R&D Spend'] = [rd_spend]
            input_data['Administration'] = [administration]
            input_data['Marketing Spend'] = [marketing_spend]

        input_data['Year'] = [year]
        input_data['Total Spend'] = [total_spend]
        input_data['Total Spend * Year'] = [total_spend_year]


        # Make prediction
        predicted_profit = loaded_model.predict(input_data)[0]

        return {
            'predicted_profit': predicted_profit,
            'R&D Spend': rd_spend,
            'Administration': administration,
            'Marketing Spend': marketing_spend,
        }
    except (ValueError, TypeError):
        return None
```
